```python
class JogoDaVelha:
    
    def __init__(self):
        
        # inicializando listas de cada jogada dos jogadores serão (usadas para marcar a jogada no tabuleiro) 
        
        self.jog_player_1 = []
        self.jog_player_2 = []
        
        # inicializando a lista de todas as jogadas dos jogadores (serão usadas para verificar o vencedor)
        
        self.lista_player_1 = []
        self.lista_player_2 = []

        # listas de jogadas para cada posição no tabuleiro
        
        self.jog_1 = ['1','1']
        self.jog_2 = ['1','2']
        self.jog_3 = ['1','3']
        self.jog_4 = ['2','1']
        self.jog_5 = ['2','2']
        self.jog_6 = ['2','3']
        self.jog_7 = ['3','1']
        self.jog_8 = ['3','2']
        self.jog_9 = ['3','3']
        
        # listas de jogadas vencedoras (se algum jogador fizer uma lista de jogadas igual a uma dessas listas será o vencedor)
        
        self.winner_1 = [['1','1'],['1','2'],['1','3']]
        self.winner_2 = [['2','1'],['2','2'],['2','3']]
        self.winner_3 = [['3','1'],['3','2'],['3','3']]
        self.winner_4 = [['1','1'],['2','1'],['3','1']]
        self.winner_5 = [['1','2'],['2','2'],['3','2']]
        self.winner_6 = [['1','3'],['2','3'],['3','3']]
        self.winner_7 = [['1','1'],['2','2'],['3','3']]
        self.winner_8 = [['1','3'],['2','2'],['3','1']]
        
    def tabuleiro(self): # método de impressão do tabuleiro
        
        print(f'+{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}+'.center(100))
        print(f'|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|'.center(100))
        print(f'|{self.x}{self.a}{self.x}|{self.x}{self.b}{self.x}|{self.x}{self.c}{self.x}|'.center(100))
        print(f'|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|'.center(100))
        print(f'|{self.z}{self.z}{self.z}|{self.z}{self.z}{self.z}|{self.z}{self.z}{self.z}|'.center(100))
        print(f'|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|'.center(100))
        print(f'|{self.x}{self.d}{self.x}|{self.x}{self.e}{self.x}|{self.x}{self.f}{self.x}|'.center(100))
        print(f'|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|'.center(100))
        print(f'|{self.z}{self.z}{self.z}|{self.z}{self.z}{self.z}|{self.z}{self.z}{self.z}|'.center(100))
        print(f'|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|'.center(100))
        print(f'|{self.x}{self.g}{self.x}|{self.x}{self.h}{self.x}|{self.x}{self.i}{self.x}|'.center(100))
        print(f'|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|{self.x}{self.x}{self.x}|'.center(100))
        print(f'+{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}{self.z}+\n'.center(100))
        
    def entrada(self, player, jog_other): # método para a escolha da jogada pelos jogadores

        while True:
            linha = input(f'{player} escolha uma linha 1, 2 ou 3:')
            
            if linha.isnumeric():
                col = input(f'{player} escolha uma coluna 1, 2 ou 3:')
                
                if col.isnumeric():
                    jogada = [linha, col] # armazenar apenas a última jogada 
                    
                    if jogada == jog_other: # esse laço serve para que o jogador não repita uma jogada feita anteriormente
                        print('ERRO! Essa posição já foi ocupada, jogue novamente\n')
                        self.entrada(player, jog_other) 
                        
                    else:
                        return jogada
                    
                else:
                    print(f'Erro! Digite um número inteiro, 1, 2 ou 3\n')
                    self.entrada(player, jog_other)
                    
            else:
                print(f'Erro! Digite um número inteiro, 1, 2 ou 3\n')
                self.entrada(player, jog_other)
    
            
    def winner_ver(self, analise): # método para a verificação se há um vencedor a cada jogada
    
        final1 = [elemento for elemento in analise if elemento in winner_1]   
        final2 = [elemento for elemento in analise if elemento in winner_2]   
        final3 = [elemento for elemento in analise if elemento in winner_3]
        final4 = [elemento for elemento in analise if elemento in winner_4]    
        final5 = [elemento for elemento in analise if elemento in winner_5] 
        final6 = [elemento for elemento in analise if elemento in winner_6]    
        final7 = [elemento for elemento in analise if elemento in winner_7]
        final8 = [elemento for elemento in analise if elemento in winner_8]    

        if final1 == winner_1:
            return True

        elif final2 == winner_2:
            return True

        elif final3 == winner_3:
            return True

        elif final4 == winner_4:
            return True

        elif final5 == winner_5:
            return True

        elif final6 == winner_6:
            return True

        elif final7 == winner_7:
            return True

        elif final8 == winner_8:
            return True

        else:
            return False
    
    def marker(self, jog_player, player): # método de marcação de jogadas no tabuleiro
        
        if jog_player == self.jog_1: 
            self.a = player
        elif jog_player == self.jog_2:
            self.b = player
        elif jog_player == self.jog_3:
            self.c = player
        elif jog_player == self.jog_4:
            self.d = player
        elif jog_player == self.jog_5:
            self.e = player
        elif jog_player == self.jog_6:
            self.f = player
        elif jog_player == self.jog_7:
            self.g = player
        elif jog_player == self.jog_8:
            self.h = player
        elif jog_player == self.jog_9:
            self.i = player
    
    
    def start(self): # método de inicialização do jogo
        
        # escolhendo o nome de cada jogador
        
        self.player_1 = input('Digite o nome do jogador 1')
        self.player_2 = input('Digite o nome do jogador 2')
        
        # inicializando as posições do tabuleiro vazias 
        
        self.a = ' '
        self.b = ' '
        self.c = ' '
        self.d = ' '
        self.e = ' '
        self.f = ' '
        self.g = ' '
        self.h = ' '
        self.i = ' '
        self.x = ' '
        self.z = '-'
        
        count = 0
        
        print('='*100)
        print('Jogo da Velha\n'.center(100))
        print('Começou!'.center(100))
        print('='*100)
        print('\n')
        
        self.tabuleiro()

        while True:
            
            self.jog_player_1 = self.entrada(self.player_1, self.jog_player_2)

            self.lista_player_1.append(self.jog_player_1) # inclui as jogadas do jogador 1 na lista para verificar resultado
            self.lista_player_1 = sorted(self.lista_player_1) # para que a lista funcione em qualquer ordem de jogada
            
            self.marker(self.jog_player_1, 'O')
            
            self.tabuleiro()

            resultado = self.winner_ver(self.lista_player_1)

            if resultado: # controle de fluxo para detreminar se jogador 1 venceu
                print('='*100)
                print(f'{self.player_1} venceu, Parabéns!\n'.center(100))
                break

            count += 1    

            if count == 9: # após 9 rodadas sem vencedor será considerado empate
                print('='*100)
                print(f'O jogo empatou!'.center(100))
                break
                
            self.jog_player_2 = self.entrada(self.player_2, self.jog_player_1)

            self.lista_player_2.append(self.jog_player_2) # inclui as jogadas do jogador 2 na lista para verificar resultado
            self.lista_player_2 = sorted(self.lista_player_2) # para que a lista funcione em qualquer ordem de jogada

            self.marker(self.jog_player_2, 'X')
            
            self.tabuleiro()

            resultado = self.winner_ver(self.lista_player_2)

            if resultado: # controle de fluxo para detreminar se jogador 2 venceu
                print('='*100)
                print(f'{self.player_2} venceu, Parabéns!\n'.center(100))
                break

            count += 1      

        print(f'Fim do Jogo'.center(100))
        print('='*100)
```

```python
j1 = JogoDaVelha()
```

```python
j1.start()
```
